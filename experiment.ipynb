```python
def get_para(link):
    response = requests.get(link)
    soup = BeautifulSoup(response.content,'html5lib')
    para = soup.find_all('p')

    cleand = []
    for p in para:
        for unwanted in p(['a', 'br', 'script', 'style']):
            unwanted.decompose()
        text = p.get_text(separator=" ", strip=True)
        cleand.append(text)
    return "\n\n".join(cleand)
        
        
```

```python
def get_data(query):
    response = search_engine(query,api_key,search_engine_id,3)
    data = []
    for r in response:
        url = r['link']
        temp = get_para(url)
        data.append(temp)
    return data
    
```
